```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Input datasets
x1 = [10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0]

y1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]

x2 = [10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0]

y2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]

x3 = [10.0, 8.0, 13.0, 9.0, 11.0, 14.0, 6.0, 4.0, 12.0, 7.0, 5.0]

y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]

x4 = [8.0, 8.0, 8.0, 8.0, 8.0, 8.0, 8.0, 19.0, 8.0, 8.0, 8.0]

y4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]
```

```python
#Calculate mean of each Dataset

X1_Mean=np.mean(x1)
X2_Mean=np.mean(x2)
X3_Mean=np.mean(x3)
X4_Mean=np.mean(x4)
Y1_Mean=np.mean(y1)
Y2_Mean=np.mean(y2)
Y3_Mean=np.mean(y3)
Y4_Mean=np.mean(y4)

Mean=[X1_Mean,X2_Mean,X3_Mean,X4_Mean,Y1_Mean,Y2_Mean,Y3_Mean,Y4_Mean]

print(Mean)
```

```python
#Calculate Variance of Datasets
X1_Var=np.var(x1)
X2_Var=np.var(x2)
X3_Var=np.var(x3)
X4_Var=np.var(x4)
Y1_Var=np.var(y1)
Y2_Var=np.var(y2)
Y3_Var=np.var (y3)
Y4_Var=np.var(y4)

Variance= [X1_Var,X2_Var,X3_Var,X4_Var,Y1_Var,Y2_Var,Y3_Var,Y4_Var]

print(Variance)
```

```python
#Calculate Corelation coefficients of Datastets
CC1=np.corrcoef(x1,y1)

CC2=np.corrcoef(x2,y2)

CC3=np.corrcoef(x3,y3)

CC4=np.corrcoef(x4,y4)

CorrCoeff=[CC1[1,0],CC2[1,0],CC3[1,0],CC4[1,0],CC1[1,0],CC2[1,0],CC3[1,0],CC4[1,0]]
print (CorrCoeff)
```

```python
#other Statistics 

#Calculate Median of each Dataset

X1_Median=np.median(x1)
X2_Median=np.median(x2)
X3_Median=np.median(x3)
X4_Median=np.median(x4)
Y1_Median=np.median(y1)
Y2_Median=np.median(y2)
Y3_Median=np.median(y3)
Y4_Median=np.median(y4)

Median=[X1_Median,X2_Median,X3_Median,X4_Median,Y1_Median,Y2_Median,Y3_Median,Y4_Median]

print(Median)



```

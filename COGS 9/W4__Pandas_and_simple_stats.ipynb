### finding rows by a value
Let's say we want to find all the movies in the database that were made in 1960. We can do so very easily by using the == operator and the panda specific loc operator.

```python
# first, let's see what == does:
df['year']==1960
```

```python
# now let's combine that with df.loc[]
df.loc[df['year']==1960]
```

```python
df_1960 = df.loc[df['year']==1960]
```

```python
df_1960.loc[126,'year']
```

```python
df_1960.iloc[0,1]
```

### now let's look at the distribution of the votes from all the movies from 1960
To do this, we need to import a plotting library called matplotlib. Don't worry about the details for now, they're all being handled by pandas in the background. As you get more familiar with python, you can customize your plots in a variety of ways with matplotlib.

```python
# the next two lines imports matplotlib and 
# tells jupyter to display the plot inline
import matplotlib.pyplot as plt
%matplotlib inline
df.loc[df['year']==1960].hist('votes',bins=50)

#perhaps try different numbers of bins?
#Or change the x values we can see by writing "plt.xlim([0,100000])" without the ""
```

### finally, let's calculate some descriptive statistics. 
Again, pandas makes it super easy. Simply attach at the very end the functions .mean() {or median, std}

```python
df.loc[df['year']==1960]['votes'].mean()
```

### and if we want to do this for all the rows for every year?

```python
df.groupby('year').mean()
```

```python
df.groupby('year')['votes'].mean().plot()
```

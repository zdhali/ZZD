```python
def count_gpr_and(string):
    if type(string)==str:
        x=re.findall("and", string)
        if len(x)!=0:
            output=len(x)
        else:
            output=0
    return output
```

```python
def count_gpr_or(string):
    if type(string)==str:
        x=re.findall("or", string)
        if len(x)!=0:
            output=len(x)
        else:
            output=0
    return output
```

```python
from_model["GPR_AND_CT"]=from_model["Gene_Reaction_Rule"].apply(count_gpr_and)
from_model["GPR_OR_CT"]=from_model["Gene_Reaction_Rule"].apply(count_gpr_or)
```

```python
#pull number of unique genes count
def count_unique_gene(string):
    unique_list=[]
    if type(string)==str:
        x=re.findall("[0-9A-Z-_.]*", string)
        if len(x)!=0:
            for i in x:
                if i!="" and i not in unique_list:
                    unique_list.append(i)
            output=len(unique_list)
        else:
            output=0
    return output
```

```python
from_model["GPR_unique_genes"]=from_model["Gene_Reaction_Rule"].apply(count_unique_gene)
```

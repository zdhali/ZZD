```python
def pull_rxn_met_(model):
    #all values must be per reaction here
    rxn_id=[]
    rxn_str=[]
    rxn_metabolites=[]
    rxn_bigg_id=[]
    rxn_gene=[]
    rxn_species=[]
    for i in model.reactions:
        rxn_id.append(i.id)
        rxn_str.append(i.reaction)
        rxn_metabolites.append(i.metabolites)
        rxn_bigg_id.append(i.notes)
        rxn_gene.append(i.gene_reaction_rule)
        rxn_species.append(str(model))
    return rxn_species, rxn_id, rxn_str, rxn_metabolites, rxn_bigg_id, rxn_gene
    
```

```python
rxn_species, rxn_id, rxn_str, rxn_metabolites, rxn_bigg_id, rxn_gene=pull_rxn_met_(iJO1366)
```

```python
df= pd.DataFrame({"Model": rxn_species, "Rxn_id": rxn_id, "Rxn_metabolites": rxn_metabolites,
                 "rxn_bigg_id": rxn_bigg_id, "rxn_genes" : rxn_gene})
df
```
